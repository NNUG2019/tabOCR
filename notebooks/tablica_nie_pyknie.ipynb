```
# funkcja z internetu
def pad_up_to(t, max_in_dims, constant_values):
    s = tf.shape(t)
    paddings = [[0, m-s[i]] for (i,m) in enumerate(max_in_dims)]
    return tf.pad(t, paddings, 'CONSTANT', constant_values=constant_values)
```

```
del(a_set)
del(b_set)
del(c_set)
```

```
f = tf.io.read_file('./test_subjects2/a' + str(3) + '.png')
print(len(tf.image.decode_png(f)))
```

```
a_set = np.array([pad_up_to(tf.image.decode_png(tf.io.read_file('./test_subjects2/a' + str(0) + '.png')), [3000, 3000, 3], 255).numpy()])
b_set = np.array([pad_up_to(tf.image.decode_png(tf.io.read_file('./test_subjects2/b' + str(0) + '.png')), [3000, 3000, 3], 255).numpy()])
c_set = np.array([pad_up_to(tf.image.decode_png(tf.io.read_file('./test_subjects2/c' + str(0) + '.png')), [3000, 3000, 3], 255).numpy()])
for i in range(1, 500):
  print('\r'+ str(i))
  f = tf.io.read_file('./test_subjects2/a' + str(i) + '.png')
  a_set = np.append(a_set, [pad_up_to(tf.image.decode_png(f), [3000, 3000, 3], 255).numpy()], axis=3)
  f = tf.io.read_file('./test_subjects2/b' + str(i) + '.png')
  b_set = np.append(b_set, [pad_up_to(tf.image.decode_png(f), [3000, 3000, 3], 255).numpy()], axis=3)
  f = tf.io.read_file('./test_subjects2/c' + str(i) + '.png')
  c_set = np.append(c_set, [pad_up_to(tf.image.decode_png(f), [3000, 3000, 3], 255).numpy()], axis=3)
```

```
a_set = tf.constant(a_set)
b_set = tf.constant(b_set)
c_set = tf.constant(c_set)
```
